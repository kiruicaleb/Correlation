Importing all the neccesary libraries and packages to help in the processes

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
data = pd.read_csv("world_cup_matches.csv")
```

```python
data.info()
```

```python
data.head()
```

```python
data.isna().sum()
```

Dropping the columns "Win Conditions" and "Date" since it has null values and will not be useful in the processing

```python
cup = data.dropna()
```

```python
cup.isna().sum()
```

```python
cup.isna().sum()
```

Creation of pairplot axes that represent all the variables in the data frame

```python
sns.pairplot(data=cup)
plt.show()
```

Narrowing down the analysis to specific columns including "Year", "Home Goals" and "Away Goals"

```python
sns.pairplot(data=cup, vars=["Year","Home Goals", "Away Goals"])
plt.show()
```

Use Kernel density estimate plots to inspect the intensity of the variables.

```python
start_year=1930
end_year=2018
filtered = cup[(cup['Year'] >= start_year) & (cup['Year'] <= end_year)]
sns.kdeplot(data=filtered, x="Year", hue="Home Goals")
plt.show()
```

```python
sns.kdeplot(data=filtered, x="Year", hue="Away Goals")
plt.show()
```

Analysis of relation plots and kernel density reveals a substantial difference between average home and away team goals. Home team scores significantly exceed those of the away team, indicating a notable advantage for home teams in scoring. This trend is evident across various datasets, suggesting a consistent pattern of higher goal scoring by home teams compared to away teams.This is imperative that the number of fans for teams most likely deteremine the success of the teams in World Cups.

```python
sns.relplot(data=cup, x="Year", y="Home Goals", hue="Home Goals")
sns.relplot(data=cup, x="Year", y="Away Goals", hue="Away Goals")
plt.show()
```

To get a more figurative visualization of the comparison between Home Team goals and Away Team goals the indroduction of a 3-Dimensional to show the space occupied by observations. 

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(cup['Year'], cup['Home Goals'], cup['Away Goals'], c='blue', marker='o')
ax.set_xlabel('Year')
ax.set_ylabel('Home Goals')
ax.set_zlabel('Away Goals')
ax.set_title('3D Scatter Plot of Home and Away Goals over Time')
```
